```python
def waktu_per_detik(t):
  return t.hour * 3600 + t.minute * 60 + t.second

df['waktu_per_detik_tap_in'] = df['tapInTime'].apply(waktu_per_detik)
df['waktu_per_detik_tap_out'] = df['tapOutTime'].apply(waktu_per_detik)\

df['waktu_perbedaan_tap_in_tap_out'] = df['waktu_per_detik_tap_out'] - df['waktu_per_detik_tap_in']
```

```python
# membuat adjacency list untuk setiap koridor
adjacency_list = {}

# membuat set untuk menyimpan nama koridor yang unik
unique_corridors = set(df['corridorName'])

# mengisi adjacency list dengan koridor sebagai kunci
for corridor in unique_corridors:
    adjacency_list[corridor] = set()

# mengisi adjacency list dengan tapOutStopsName yang relevan
for index, row in df.iterrows():
    corridor = row['corridorName']
    tap_out_stop = row['tapOutStopsName']
    if pd.notna(corridor) and pd.notna(tap_out_stop): # Check for non-null values
        adjacency_list[corridor].add(tap_out_stop)

# Menampilkan adjacency list (opsional)
for corridor, stops in adjacency_list.items():
     print(f"{corridor}: {stops}")
```

```python
import json
json_output = json.dumps({k: list(v) for k, v in adjacency_list.items()}, indent=4)
# You can now print this JSON string
print(json_output)
# Or, you can save it to a JSON file
with open('adjacency_list.json', 'w') as f:
    json.dump({k: list(v) for k, v in adjacency_list.items()}, f, indent=4)
print("\nAdjacency list saved to adjacency_list.json")
```

```python
def build_ltree(adjacency_list, start_node):
    def dfs(node, path):
        current_path = path + "." + node if path else node
        children = sorted(list(adjacency_list.get(node, set())))
        if not children:
            return current_path
        else:
            return current_path + ".*{" + ",".join(dfs(child, current_path) for child in children) + "}"

    if start_node not in adjacency_list:
        return start_node

    # Start the DFS without an initial path prefix
    return dfs(start_node, "")

if adjacency_list:
    starting_node = list(adjacency_list.keys())[0]
    ltree_representation = build_ltree(adjacency_list, starting_node)
    print(f"\nLtree representation starting from '{starting_node}':")
    print(ltree_representation)
else:
    print("Adjacency list is empty. Cannot build Ltree.")
```
